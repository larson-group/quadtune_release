```python
ds = params_dataset
```

```python
obs_values = ds[["SWCF","LWCF"]].isel(time=0,product=1,ens_idx=0)
weights = ds["area"].values
```

```python
nrmlzdWeights =weights/np.sum(weights)
np.sum(nrmlzdWeights * obs_values)


obs_values.values
```

```python
ds_filtered = ds[['SWCF','params','LWCF']]


new_ds = xr.Dataset({'params': ds_filtered['params']})


lat_size = ds_filtered.sizes['lat']
lon_size = ds_filtered.sizes['lon']
fields = ["SWCF","LWCF"]
for idx,field in enumerate(fields):
    for i in range(1,lat_size+1):
        for j in range(lon_size):
            metric_name = f'{field}_{i}_{j+1}'
            cell_data = ds_filtered[field].isel(lat=-i, lon=j).drop_vars(['lat', 'lon'])
            new_ds[metric_name] = cell_data

            if idx == 0:    
                metric_weight_name = f"numb_{i}_{j+1}"
                new_ds[metric_weight_name] = nrmlzdWeights[-i, j]
```
